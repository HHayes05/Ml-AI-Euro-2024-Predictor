```python
import pandas as pd
```

```python
df_historical_data = pd.read_csv('Euro_results.csv')
df_fixture= pd.read_csv('euro_fixture.csv')
```

```python
df_fixture['Home Team'] = df_fixture['Home Team'].str.strip()
df_fixture['Away Team'] = df_fixture['Away Team'].str.strip()
df_historical_data
```

```python
df_historical_data.drop_duplicates(inplace=True)
df_historical_data.sort_values('Year', inplace=True)
df_historical_data
```

```python
df_historical_data['Game Score'] = df_historical_data['Game Score'].str.replace('[^\d–]', '', regex=True)
```

```python
df_historical_data['Home Team'] = df_historical_data['Home Team'].str.strip()
df_historical_data['Away Team'] = df_historical_data['Away Team'].str.strip()
```

```python
df_historical_data[['Home Goals', 'Away Goals']] = df_historical_data['Game Score'].str.split('–', expand=True)
df_historical_data.drop('Game Score', axis=1, inplace=True)
df_historical_data = df_historical_data.astype({'Home Goals': int, 'Away Goals':int, 'Year': int})

# creating new column "totalgoals"
df_historical_data['Total Goals'] = df_historical_data['Home Goals'] + df_historical_data['Away Goals']
df_historical_data
```

```python
df_historical_data.to_csv('cleaned_euro_results.csv', index=False)
```

```python
df_fixture.to_csv('cleaned_euro_fixture.csv', index=False)
```
